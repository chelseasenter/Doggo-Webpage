```python
import pandas as pd
```

```python
dogs_df = pd.read_csv("graph_dog.csv")
```

```python
del dogs_df["Unnamed: 0"]
dogs_df.head(3)
```

```python
dogs_df[["min_weight", "max_weight"]] = dogs_df["Weight"].str.split("-",expand = True)
dogs_df.head()
```

```python
dogs_df[["min_height", "max_height"]] = dogs_df["Height"].str.split("-",expand = True)
dogs_df.head()
```

```python
import numpy as np
from itertools import chain

# return list from series of comma-separated strings
def chainer(s):
    return list(chain.from_iterable(s.str.split(',')))

# calculate lengths of splits
lens = dogs_df['Temperament'].str.split(',').map(len)

# create new dataframe, repeating or chaining as appropriate
res = pd.DataFrame({'Dog Breed': np.repeat(dogs_df['Dog Breed'], lens),
                    'Breed Group': np.repeat(dogs_df['Breed Group'], lens),
                    'Temperament': chainer(dogs_df['Temperament'])})

print(res)
```

```python
res.to_csv('sunburst.csv', index = False)
```

```python
import plotly.express as px
fig = px.sunburst(res, names = "Temperament", parents = "Dog Breed", values = "Breed Group")
fig.show()
```

```python
import plotly.express as px
import numpy as np

fig = px.sunburst(res, path=['Breed Group', 'Dog Breed'])
fig.show()
```
