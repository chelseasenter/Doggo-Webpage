```python
#import dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress
```

```python
#import data
dogs_df = pd.read_csv("graph_dog.csv")
```

```python
#get rid of extraneous column
del dogs_df["Unnamed: 0"]
dogs_df.head(3)
```

```python
#split weight into single entries
dogs_df[["min_weight", "max_weight"]] = dogs_df["Weight"].str.split("-",expand = True)
dogs_df.head()
```

```python
#split height into single entries 
dogs_df[["min_height", "max_height"]] = dogs_df["Height"].str.split("-",expand = True)
dogs_df.head()
```

```python
#delete word "years" from age range
dogs_df['Life Span'] = dogs_df['Life Span'].str.replace('years', '')
#delete word "Years" from age range
dogs_df['Life Span'] = dogs_df['Life Span'].str.replace(' Years ', '')

dogs_df.head()
```

```python
#split life span to min max 
dogs_df[["min_lifespan", "max_lifespan"]] = dogs_df["Life Span"].str.split("-",expand = True)
dogs_df.head()
```

```python
#convert max weight to float
dogs_df["max_weight"] = dogs_df.max_weight.astype(float)
dogs_df.head()
```

```python
#convert max height to float
dogs_df["max_height"] = dogs_df.max_height.astype(float)
dogs_df.head()
```

```python
#convert max life span to float
dogs_df["max_lifespan"] = dogs_df.max_lifespan.astype(float, errors= 'ignore')
#drop NaN
dogs_df.dropna()
dogs_df.head()
```

```python
dogs_df.dtypes
```

```python
#Lifespan vs Weight plot
x_values = dogs_df["max_weight"]
y_values = dogs_df["max_lifespan"]
mask = ~np.isnan(x_values) & ~np.isnan(y_values)
(slope, intercept, r_value, p_value, std_err) = linregress(x_values[mask], y_values[mask])

regress_values = x_values * slope + intercept
plt.scatter(dogs_df["max_weight"],dogs_df["max_lifespan"], c='y', alpha = 0.5)
plt.plot(x_values,regress_values,"-")
r_weight = round(r_value**2,2)
plt.annotate(f'r squared = {r_weight}',(10,10),fontsize=12,color="black")
plt.xlabel("Weight in Pounds")
plt.ylabel("Lifespan in Years")
plt.title("Lifespan Compared to Weight")

print(f"The r-squared is: {r_weight}")
#save as png
plt.savefig('lifespanvsweight.png')
plt.show()

```

```python
#Lifespan vs Height plot
x_values = dogs_df["max_height"]
y_values = dogs_df["max_lifespan"]
mask = ~np.isnan(x_values) & ~np.isnan(y_values)
(slope, intercept, r_value, p_value, std_err) = linregress(x_values[mask], y_values[mask])

regress_values = x_values * slope + intercept
plt.scatter(dogs_df["max_height"],dogs_df["max_lifespan"], c='b', alpha = 0.5)
plt.plot(x_values,regress_values,"-")
r_height = round(r_value**2,2)
plt.annotate(f'r squared = {r_height}',(10,10),fontsize=12,color="black")
plt.xlabel("Height in Inches")
plt.ylabel("Lifespan in Years")
plt.title("Lifespan Compared to Height")

print(f"The r-squared is: {r_height}")

#save as png
plt.savefig('lifespanvsheight.png')
plt.show()
```

```python
# import temperament count from csv
temperament_df = pd.read_csv('temperamentcount.csv')
temperament_df = temperament_df.rename(columns = {"Unnamed: 0":"Temperament"})
temperament_df.head()
```

```python
#Lifespan vs Weight plot
plt.figure(figsize=(20,9))
plt.bar(temperament_df['Temperament'],temperament_df['Temperament Count'], alpha = 0.5, color=plt.cm.Paired(np.arange(len(temperament_df['Temperament']))))
x_axis = np.arange(len(temperament_df['Temperament']))
tick_locations = [value for value in x_axis]
plt.xticks(tick_locations, temperament_df['Temperament'], rotation = 'vertical')
plt.xlabel("Temperament")
plt.ylabel("Number of Breeds")
plt.title("Number of Breeds with Certain Temperaments")
plt.margins(0.01)
#save as png
plt.savefig('temperamentcount.png')
plt.show()
```

```python
import numpy as np
from itertools import chain

# return list from series of comma-separated strings
def chainer(s):
    return list(chain.from_iterable(s.str.split(',')))

# calculate lengths of splits
lens = dogs_df['Temperament'].str.split(',').map(len)

# create new dataframe, repeating or chaining as appropriate
res = pd.DataFrame({'Dog Breed': np.repeat(dogs_df['Dog Breed'], lens),
                    'Breed Group': np.repeat(dogs_df['Breed Group'], lens),
                    'Temperament': chainer(dogs_df['Temperament'])})

print(res)
```

```python
res.to_csv('sunburst.csv', index = False)
```

```python
# Python program to generate WordCloud
  
# importing all necessery modules
from wordcloud import WordCloud, STOPWORDS

  
comment_words = ''
stopwords = set(STOPWORDS)
  
# iterate through the csv file
for val in res['Temperament']:
      
    # typecaste each val to string
    val = str(val)
  
    # split the value
    tokens = val.split()
      
    # Converts each token into lowercase
    for i in range(len(tokens)):
        tokens[i] = tokens[i].lower()
      
    comment_words += " ".join(tokens)+" "
  
wordcloud = WordCloud(width = 800, height = 800,
                background_color ='white',
                stopwords = stopwords,
                min_font_size = 10).generate(comment_words)
  
# plot the WordCloud image                       
plt.figure(figsize = (8, 8), facecolor = None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad = 0)
plt.savefig("wordcoudtemperaments.png")
plt.show()
```
